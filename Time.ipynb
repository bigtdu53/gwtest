```python
def mean_time(X,clf,N=10):
    elapsed=[]
    start_time = time.time()
    for i in range(N):
        n1=random.randint(0,len(X)-1)
        n2=random.randint(0,len(X)-1)
        sim=clf.compute_similarity(X[n1],X[n2])
    elapsed.append(time.time() - start_time)
    return np.mean(elapsed)
```
